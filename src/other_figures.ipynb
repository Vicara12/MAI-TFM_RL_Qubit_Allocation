```python
from utils.customtypes import Circuit
from sampler.randomcircuit import RandomCircuit, DenseRandomCircuit, HotRandomCircuit
from utils.plotter import drawCircuit
```

# Random Circuit Sampling methods

```python
n_qubits = 6
random_circuit =      RandomCircuit(num_lq=n_qubits, num_slices=n_qubits).sample()
hot_circuit    =   HotRandomCircuit(num_lq=n_qubits, num_slices=n_qubits).sample()
dense_circuit  = DenseRandomCircuit(num_lq=n_qubits, num_slices=n_qubits).sample()

drawCircuit(circuit_slice_gates=random_circuit.slice_gates, num_lq=n_qubits, title='Random Circuit')
drawCircuit(circuit_slice_gates=hot_circuit.slice_gates,    num_lq=n_qubits, title='Hot Circuit')
drawCircuit(circuit_slice_gates=dense_circuit.slice_gates,  num_lq=n_qubits, title='Dense Circuit')
```
